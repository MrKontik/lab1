```python
def blowout(att):
    x = data.sort_values(by=att)
    x=x[att]
    x25=x[24]
    x75=x[74]
    start=x25-1.5*(x75-x25)
    end=x75+1.5*(x75-x25)
    average=x.mean()
    x[(x<start) & (x>end)]=average
    print("Было среднее")
    print(data[att].mean())
    print("Стало среднее")
    print(x.mean())
    print("Была дисперсия")
    print(data[att].std()*data[att].std())
    print("Стала дисперсия")
    print(x.std()*x.std())
    print("Нормировка")
    print((x-x.mean())/(x.std()*x.std()))
    
```
